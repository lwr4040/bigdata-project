```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

# Image Load
image = io.imread('/content/drive/MyDrive/whiteCar.JPG')
image2= cv2.imread('/content/drive/MyDrive/Screenshot_20220329-130253_Zoom.jpg')
# Grayscale
gray_image = np.float32(cv2.cvtColor(image, cv2.COLOR_RGB2GRAY))

embossing = np.array([[-1, -1, 0], [-1, 0, 1], [0, 1, 1]])

# Embossing     pixel  +128
output = cv2.filter2D(gray_image, -1, embossing) + 128

# uint8 
output[output > 255] = 255
output[output < 0] = 0
output = np.uint8(output)

height ,width =output.shape
matrix= cv2.getRotationMatrix2D((width/2, height/2),135,0.5)
dst= cv2.warpAffine(output, matrix, (width,height))
im= cv2.resize(image2,(4032,3024))
love=cv2.hconcat([output,dst])
love=cv2.resize(love,(320,300))
cv2_imshow(love)
```

````r
team_avg <- DF %>% group_by(team, year) %>% summarise(avg_AVG= mean(avg, na.rm = T)) %>%
                filter(year=="2016"|team=='KIA')
team_avg %>% show()
````

````r
circle <- table(team_avg$avg_AVG>=0.3)
pct<-round(circle/sum(circle)*100,2)
label<-paste(names(circle),"\n",c("3할 이하","3할 이상"),pct,"%")
pie(circle,main="KIA의 2016년 타율",xlab="team",ylab="batter_id",labels=label)
# 2016년에 3할 이상 치는 타자의 비중이 15.38%이었고, 나머지 84.62%는 타율이 3할이 넘지 못했다.
# 3할 이상치는 타자는 매우 잘 친다라는 것을 알 수 있다. 
````

````r
team_year_avg <- DF %>% group_by(team, year) %>% summarise(avg_AVG = mean(avg, na.rm = T)) %>%
                      filter(team =="NC")
team_year_avg %>% show()
````

````r
barplot(team_year_avg$avg_AVG,names.arg=team_year_avg$year,main="NC의 연도별 타율",xlab="year",ylab="avg_AVG",col=c("red","orange","yellow","dark green","blue","purple"))
# barplot으로 나타낸 다음, NC의 연도별 타율을 구해 연도별로 타율이 얼마나 올랐는지 시각화 한다.
# 최고점은 2할 5푼, 최저점은 2할 3푼이라는 것을 그래프를 통해 확인할 수 있다.
````

````r
batter<-team_year_avg$avg_AVG
histo<-hist(batter,main="NC의 연도별 타율",xlab="avg_AVG",ylab="Frequency",col="green",ylim=c(0,6))
#2할 2푼대와 2할 3푼대가 3번으로 가장 많았고, 2할 4푼과 2할 5푼인적은 단 한번도 있지 않았다는 것을 알 수 있다.
#NC의 팀 타율이 상대적으로 중간정도였다는 것을 알 수 있다.
````
